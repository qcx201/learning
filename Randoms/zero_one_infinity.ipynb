```python
class One():
    
    def __init__(self):
        pass
    
    def __repr__(self):
        return 'One'
    
    def __add__(self, other):
        return other + 1
    
    def __radd__(self, other):
        return self.__add__(other)
    
    def __sub__(self, other):
        return -other + self
    
    def __rsub__(self, other):
        return other + (-self)
    
    def __neg__(self):
        return NegOne()
    
    def __mul__(self, other):
        return other
    
    def __rmul__(self, other):
        return other
    
    def __rtruediv__(self, other):
        return "Fractions"
        
class NegOne():
    
    def __init__(self):
        pass
    
    def __repr__(self):
        return 'NegOne'
    
    def __add__(self, other):
        return One() - other
    
    def __sub__(self, other):
        return -(other + One())
    
    def __rsub__(self, other):
        return other + One()
    
```

```python
class Infty(object):
    
    def __init__(self):
        pass
    def __repr__(self):
        return '∞'
    
    def __add__(self, other):
        return self
    
    def __radd__(self, other):
        return self + other

    def __neg__(self):
        return NegInfty()
    
    def __sub__(self, other):
        return self
    
    def __rsub__(self, other):
        return other + (-self)
    
    def __mul__(self, other):
        return self
    
    def __rmul__(self, other):
        return self
    
    def __truediv__(self, other):
        return self
    
    def __rtruediv__(self, other):
        return 0
    
    def __gt__(self, other):
        return True
    
    def __ge__(self, other):
        return True
    
    def __lt__(self, other):
        return False
    
    def __le__(self, other):
        return False
    

class NegInfty(Infty):
    def __init__(self):
        pass
    def __repr__(self):
        return '-∞'
    
    def __gt__(self, other):
        return False
    
    def __ge__(self, other):
        return False
    
    def __lt__(self, other):
        return True
    
    def __le__(self, other):
        return True
```
