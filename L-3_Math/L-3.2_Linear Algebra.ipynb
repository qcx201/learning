# Chapter 2: Vectors

```python
textbu('Vectors', tag='h1')

vspace()

a, b, c, d = sp.symbols('a, b, c, d')

# a column vector
v1 = sp.Matrix([a, b])

# a row vector
v2 = sp.Matrix([a, b]).T # transpose

# a column vector
v3 = sp.Matrix([c, d])

# show
for i, v in enumerate([v1, v2, v3]):
    
    tbu(f'Vector {i}:')
    
    Lx(v).show()
    
    md('shape (rows, columns):', v.shape)
```

## Plot Vectors

```python
# origin (zero vector)
v0 = np.array([0, 0])

# first vector
v1 = np.array([0, 1])

# second vector
v2 = np.array([1, 0])


# (label, origin, vector) triples
OGV = [
    # add v2 to v1
    (v0, v1, '$v_1$'),
    (v1, v2, '$v_2$'),
    # add vectors
    (v0, v1 + v2, '$v_1 + v_2$'),
]

fig, ax, plots = plot_vectors(*transpose(OGV))
plt.show()

# show vector latex
Lx(r'\quad').join(
    Lx('v_1', v1),
    Lx('v_2',  v2),
    Lx('v_1 + v_2', v1+v2)
).show(center=False)
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

ax.quiver(0, 0, 1, 1,
          scale=1,
          width=0.0125,
          angles='xy',
          color=dcolors[0],
          scale_units='xy')

ax.quiver(1, 1, -1, 1,
          scale=1,
          width=0.0125,
          angles='xy',
          color=dcolors[1],
          scale_units='xy')

ax.set_xlim(-0.1, 3)
ax.set_ylim(-0.1, 3)

ax.grid()

plt.show()
```

```python
%matplotlib notebook

# 3-D plot
fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection='3d')

fmt = dict(arrow_length_ratio=0.15, lw=3)

ax.quiver(0, 0, 0,
          1, 1, 1,
          **fmt
         )

ax.quiver(1, 1, 1,
          1, 1, -1,
          **fmt
         )

ax.set_title('3-D Vector')

ax.set_xlim(0, 3)
ax.set_ylim(0, 3)
ax.set_zlim(0, 3)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```
