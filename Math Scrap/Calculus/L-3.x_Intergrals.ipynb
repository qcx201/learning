```python
import numpy as np
import sympy as sym

import matplotlib.pyplot as plt
%matplotlib inline
```

$$
y^2 + x^2 = r \\
y = \sqrt{1 - x^2}
$$

```python
x = sym.symbols('x')

y = sym.sqrt(1 - x**2)
res = y.integrate((x, 0, 1))

display(res)
```

```python
for n in [5, 10, 25, 50, 100, 500, 1000]:
    
    Xbar = np.linspace(0, 1, n)

    plt.figure(figsize=(8, 8))

    width = np.abs(Xbar[1] - Xbar[0])
    Ybar = np.sqrt(1 - Xbar**2)
    plt.bar(Xbar, Ybar, fill=None,
            width=width, alpha=0.5,
            label=r'$\bar{\pi} = ' + f'{4*(Ybar*width).sum():.2f}$'
           )

    plt.scatter(Xbar, Ybar, s=100, alpha=0.5)
    
    plt.legend(fontsize=18)
    
    X = np.linspace(0, 1, 1_000)
    plt.plot(X, np.sqrt(1 - X**2), lw=6, alpha=0.5)

    plt.xlim(0, 1.1)
    plt.ylim(0, 1.1)
    plt.show()
```

```python
plt.figure(figsize=(8, 8))

X = np.linspace(-1, 1, 10_000)

format_kw = dict(alpha=0.75, lw=5)

Y = np.sqrt(1 - X**2)

for y in [Y, -Y]:
    
    plt.plot(X,  y, **format_kw)

plt.show()
```
