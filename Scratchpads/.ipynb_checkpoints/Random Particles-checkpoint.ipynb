```python
import numpy as np
import random
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
n = 500

μ, σ = 0, 1 # standard normal

particles = np.array([(0.0, 0.0),] * n)

α, Δα = 1, 0.92

fig = plt.figure(figsize=(5, 5))

bounds = 15
plt.xlim((-bounds, bounds))
plt.ylim((-bounds, bounds))

for i in range(20):

    plt.scatter(particles[:,0], particles[:,1], alpha=α, facecolors='none', edgecolors='#4488dd')
    
    plt.show()

    # update
    Δ = np.random.normal(μ, σ, particles.shape)
    particles += Δ
    α *= Δα
```

```python
n = 500
μ_x, σ_x = 0.5, 5
μ_y, σ_y = 0, 1

particles = np.array([(0.0, 0.0),] * n)

α, Δα = 1, 0.92

fig = plt.figure(figsize=(5, 5))

bounds = 20
plt.xlim((-1, bounds*2))
plt.ylim((-bounds, bounds))

ims = []

for i in range(20):
    
    im = plt.scatter(particles[:,0], particles[:,1], alpha=α, facecolors='none', edgecolors='#4488dd')
    
    plt.show()
    
    # update
    ΔX = np.random.normal(μ_x, σ_x, particles.shape[0])
    ΔY = np.random.normal(μ_y, σ_y, particles.shape[0])
    Δ = np.array([ΔX, ΔY]).T
    
    particles += Δ
    α *= Δα
```
