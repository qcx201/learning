# Chapter 2: Differentiation üé©üêá

`Recall` in the last chapter we had experimented with functions of position and time, with changes of functions over time.

These are concepts related to a branch of mathematics called **`differential`** **`calculus`**, which studies the relationship between functions and change.


`Recall` our previous formulations:
$$
\begin{align}
\text{Position: } \, x(t) &= \text{some function, for example: }\, -\frac{49}{10}t^{2} + 500 \\
\\
\text{Instantaneous Velocity: } \, v(t) &= \frac{x(t+\Delta{t}) - x(t)}{\Delta{t}}, \, \text{ for some very small } \, \Delta{t}
\\
\\
\text{Instantaneous Acceleration: } \, a(t) &= \frac{v(t+\Delta{t}) - v(t)}{\Delta{t}}, \, \text{ for some very small } \, \Delta{t}
\end{align}
$$

## Some Jargon üí¨üôä

In our example, the functions for velocity `v(t)` and acceleration `a(t)` are both derived from **`x(t)`**.

We say that:

* **`v(t)`** is the ***first derivative*** of `x(t)`, or just the ***derivative***.


* **`a(t)`** is the derivative of `v(t)`, and also the ***second derivative*** of `x(t)`.


And we say that:


* We ***differentiate `x(t)`*** to obtain `v(t)`;

    Or, we ***take the derivative of `x(t)`*** to obtain `v(t)`.



* We ***differentiate `v(t)`*** to obtain `a(t)`;

    Or, we ***take the derivative of `v(t)`*** to obtain `a(t)`.
    

Let's examine the concept of differentiation more generally.

# The Linear Equation üìè

A common formulation of differentiation begins with the single-variable linear function in the familiar slope-intercept form:

$$
y = mx + b \\
$$

Where ***`(x, y)`*** forms the coordinates of a line with slope ***`m`*** and y-intercept ***`b`***, both of which are constant.

### Solution

```python
# Question 1.

def f(x):
    '''
    Some function of x
    '''
    p = 3 # power
    
    fac = 1 # factorial
    for i in range(1, p+1):
        fac*=i
        
    res = (x**p)/fac # result
    
    return res
```

Let's choose two arbitrary numbers to represent points on the $x$-axis. We can call them $x_{0}$ and $x_{1}$.

Using these two points and our function `f(x)`, we can create two coordinates on the xy-plane:

$$
\text{coordinate 1}:\quad \bigg(x_{0}, \, f(x_{0})\bigg) \\
\text{coordinate 2}:\quad \bigg(x_{1}, \, f(x_{1})\bigg) \\
$$

These two equations represent two points on a `line` on the xy-plane. We'll call the linear function the describes this line ***g(x)***, which has a `constant slope` and `y-intercept` value.

### Slope and Intercept

With the coordinates, we can solve for the `slope`, ***`m`***:

$$
\begin{align}
m &= \frac{\Delta{y}}{\Delta{x}} \\
\\
&=\frac{y_{1} - y_{0}}{x_{1} - x_{0}} \\
\\
&=\frac{f(x_{1}) - f(x_{0})}{x_{1} - x_{0}} \\
\\
&=\frac{f(x_{0} + \Delta{x}) - f(x_{0})}{\Delta{x}} \\
\end{align}
$$

And then, with `m`, solve the `intercept`, ***b***:
$$
\begin{align}
b &= y-mx \\
&= f(x_0) - mx_0 \\
&= f(x_1) - mx_1 \\
\end{align}
$$

You then have all the parameters needed to describe `g(x)`.

```python
# 1.
x0, x1 = 0, 4

# slope
delta_y = f(x1) - f(x0)
delta_x = x1 - x0

m = delta_y / delta_x

# intercept
b = f(x0) - m*x0

print(f'slope: {m:.2f}')
print(f'y-intercept: {b:.2f}')
print(f'\nEquation in slope-intercept form:', '\n----')
print(f'y = {m:.2f}x + {b:.2f}')
```


## Notation of Derivatives

And some choices for use [notation of differentiation](https://en.wikipedia.org/wiki/Notation_for_differentiation#Lagrange's_notation).

Leibniz's and Lagrange's notations are commonly used.

Newton's, less so... not sure why, maybe he should have been a little nicer to some people ü§∑‚Äç‚ôÇÔ∏è.



#### Taking our example of $x(t)$:

#### Leibniz's Notation
$$
\begin{align}
\\
1^{st} \text{derivative:}\qquad \frac{d x(t)}{dt} &= v(t) \\ 
\\
2^{nd} \text{derivative:} \qquad \frac{d^{2} x(t)}{dt^{2}} &= \frac{d v(t)}{dt} = a(t)\\ 
\\
3^{rd} \text{derivative:} \qquad \frac{d^{3} x(t)}{dt^{3}} &= \frac{d^{2} v(t)}{dt^{2}} = \frac{d a(t)}{dt} = ...\\ 
\vdots \qquad \qquad \qquad & \vdots \\
\end{align}
$$


#### Lagrange's notation
$$
\begin{align}
\\
1^{st} \text{derivative:} \qquad x'(t) &= v(t) \\ 
\\
2^{nd} \text{derivative:} \qquad x''(t) &= x^{(2)}(t) = v'(t) = a(t)\\ 
\\
3^{rd} \text{derivative:} \qquad x'''(t) &= x^{(3)}(t) = v''(t) = v^{(2)}(t) = a'(t) = ...\\ 
\vdots \qquad \qquad \qquad & \vdots \\
\end{align}
$$
