# Chapter 2: Differentiation üé©üêá

`Recall` in the last chapter we had experimented with functions of position and time, with changes of functions over time.

These are concepts related to a branch of mathematics called **`differential`** **`calculus`**, which studies the relationship between functions and change.


`Recall` our previous formulations:
$$
\begin{align}
\text{Position: } \, x(t) &= \text{some function, for example: }\, -\frac{49}{10}t^{2} + 500 \\
\\
\text{Instantaneous Velocity: } \, v(t) &= \frac{x(t+\Delta{t}) - x(t)}{\Delta{t}}, \, \text{ for some very small } \, \Delta{t}
\\
\\
\text{Instantaneous Acceleration: } \, a(t) &= \frac{v(t+\Delta{t}) - v(t)}{\Delta{t}}, \, \text{ for some very small } \, \Delta{t}
\end{align}
$$

---
## Derivative Relationships üôäüí¨

In the previous chapter, we **`derived`** both velocity `v(t)` and acceleration `a(t)` from **`x(t)`**.

We say that:

* **`v(t)`** is the ***first derivative*** of `x(t)`, or just the ***derivative***.


* **`a(t)`** is the *derivative* of `v(t)`, and also the ***second derivative*** of `x(t)`.

<br>
And, to describe their relationships in the context of differential calculus, we say that:


* We ***`differentiate`*** $x(t)$ to obtain $v(t)$;


* Or, we ***take the `derivative` of*** $x(t)$ to obtain $v(t)$;


* Or, $v(t)$ ***is the `derivative` of*** $x(t)$ .

These all mean the same thing. Following the same pattern:

* *We `differentiate` $v(t)$* to obtain $a(t)$;


* Or, we *take the `derivative` of $v(t)$* to obtain $a(t)$.


* Or, $a(t)$ *is the `derivative` of* $v(t)$.

<br>

In this chapter, we'll look at the concept of `differentiation or derivatives` more closely üîçüî¨üëã.

The goal of this chapter is to understand how to apply `differentiation` of `any` general function, that is differentiable.

# Ch 2.2: The Linear Function üìèüí¶
**... slope-intercepts got me like:**

![alt text](https://media.giphy.com/media/A4R8sdUG7G9TG/giphy.gif)

A common way to start thinking about differentiation begins with the single-*(and ready to mingle* üç∏*)*-variable linear function in the slope-intercept form:

$$
y = mx + b \\
$$

Where ***`(x, y)`*** forms the coordinates of a line with slope ***`m`*** and y-intercept ***`b`***, both of which are **constant** (`m` and `b` don't change as x and y changes).

As we'll see, we are especially interested in the ***slope, `m`***.

**Our next step** is to construct a linear function related to our arbitrary function above.

<a name="ch2.2.1"></a>
## 2.2.1: Pick two points üü¢üü°üé©
**... any two points.**



Let's choose two arbitrary numbers to represent points on the $x$-axis. We can call them $x_{0}$ and $x_{1}$.

Using these two points, with our function `f(x)` we can create two `x, y coordinates`:

$$
\text{1st coordinate}:\quad \bigg(x_{0}, \, f(x_{0})\bigg) \\
\text{2nd coordinate}:\quad \bigg(x_{1}, \, f(x_{1})\bigg) \\
$$

<a name="ch2.2.2"></a>
## 2.2.2: The secant üôà‚Üî
**Can't you see?**

These two coordinates represent two points on the xy-plane, through which you can draw a `straight line`.

This line connects $f(x_{0})$ and $f(x_{1})$, and is called the **`secant`**.

The `secant` can be described with a linear function, which has a constant `slope` and `y-intercept` value.


<a name="ch2.2.3"></a>
## 2.2.3: Slope and Intercept üèÄ‚ùåüß∫

With two coordinates, we can solve for the ***`slope`, `m`***:

$$
\begin{align}
m &= \frac{\Delta{y}}{\Delta{x}} \\
\\
&=\frac{y_{1} - y_{0}}{x_{1} - x_{0}} \\
\\
&=\frac{f(x_{1}) - f(x_{0})}{x_{1} - x_{0}} \\
\\
&=\frac{f(x_{0} + \Delta{x}) - f(x_{0})}{\Delta{x}} \\
\end{align}
$$

Then, having a value for `m`, we can solve for the ***`y-intercept`, `b`***:

$$
\begin{align}
b &= y-mx \\
&= f(x_0) - mx_0 \\
&= f(x_1) - mx_1 \\
\end{align}
$$

<br>

Now we'll have both parameters needed to describe and plot the **`secant`**.

### Solutions

```python
# Ch 2.2: Question 1.

def slope_intercept(x0, x1):
    
    # delta's
    delta_y = f(x1) - f(x0)
    delta_x = x1 - x0
    
    # slope
    m = delta_y / delta_x

    # intercept
    b = f(x0) - m*x0
    
    return m, b
```

```python
# Ch 2.2: Question 2.

def plot_secant(secant_points, plot_lim=5):
    
    '''
    Plots f(x) and secant slope for delta_x in to_zero, approaches zero
    
    Parameters:
    --
    secant_points : list or tuple
        two values used to calculate the secant 
    
    plot_lim : int or float
        magnitude of the plot's xlim; the domain of the function
    '''
    
    # x-axis range
    x_range = [i/10 for i in range(-10*plot_lim, 11*plot_lim)]
    
    # data for plotting f(x)
    f_range = [f(x) for x in x_range]

    # separate values of secant points
    x0, x1 = secant_points
    
    
    # secant parameters
    m, b = slope_intercept(x0, x1)
    
    # data for plotting secant
    secant = [m*x + b for x in x_range]

    # set figure size
    fig = plt.figure(figsize=(8, 5))

    # set axis labels and title
    plt.title('Plot of f(x) and Secant', fontsize=18, fontweight='bold')

    # set axis plot limits
    plt.xlim(-plot_lim, plot_lim)
    plt.ylim(f(-plot_lim), f(plot_lim))
    
    # axis labels
    kwargs = dict(fontsize=14, rotation=0, labelpad=25)
    plt.xlabel('$x$', **kwargs)
    plt.ylabel('$y$', **kwargs)

    # common arguments for f(x) and secant plots
    kwargs = dict(lw=5, alpha=0.55)

    # plot f(x)
    plt.plot(x_range, f_range, **kwargs)

    # plot secant
    plt.plot(x_range, secant,  **kwargs)

    
    # labeling secant points
    for x in secant_points:

        # y-coordinate of secant point
        y = f(x)
        
        # plot the point
        plt.scatter(x, y, lw=8, alpha=1, color='#FF6666', zorder=5)

        # sight lines
        kwargs = dict(ls='--', lw=2,  alpha=0.15)
        plt.vlines(x, f_range[0], y, **kwargs)
        plt.hlines(y, x_range[0], x, **kwargs)

        # coordinate values
        plt.text(x, -f(plot_lim)*1.15, f'{x:.3f}',ha='center', va='top',  alpha=0.5)
        plt.text(-plot_lim*1.1, y, f'{y:.3f}', ha='right', va='center', alpha=0.5)
        

        # length of "feet" for showing delta_x and delta_y
        limb = 1
        plt.vlines(x, -f(plot_lim)*0.9-limb, -f(plot_lim)*0.9+limb, alpha=0.9)
        plt.hlines(y, -plot_lim*0.9-(limb/10), -plot_lim*0.9+(limb/10), alpha=0.9)

    # show delta_x
    delta_x = x1 - x0
    plt.hlines(-f(plot_lim)*0.9, x0, x1, alpha=0.9)
    plt.text(x0+(delta_x/2), -f(plot_lim)*0.8, '$\Delta{x}: '+f'{delta_x:.5f}$', fontsize=12, ha='center')

    # show delta_y
    delta_y = f(x1) - f(x0)
    plt.vlines(-plot_lim*0.9, f(x0), f(x1), alpha=0.9)
    plt.text(-plot_lim*0.85, f(x1)-delta_y/2, '$\Delta{y}: '+f'{delta_y:.5f}$', fontsize=12, va='bottom', ha='left')

    # show slope
    plt.text((x0+delta_x/2), (f(x0)+delta_y/2)+5, f'slope: {m:.5f}', fontsize=14, va='center', ha='left')

    
# plot result
plot_secant((0,4))

plt.show()


print('\nüëç')
```

# Ch 2.4: Infinity and Infinitesimals üöÄüåå

As we should see with the questions above, at a point $x$, and as $\Delta{x} \to 0$, the slope of the secant $\frac{f(x+\Delta{x}) - f(\Delta{x})}{\Delta{x}}$ approaches some value.

<br>
This result may see arbitrary at first, but it is actually pretty interesting since we have $\Delta{x}$ in the denominator.

You may wonder that, as $\Delta{x} \to 0$, doesn't $\frac{1}{\Delta{x}} \to \infty$? If you did, you would be correct! üôåü•≥

<a name="ch2.4.1"></a>
## 2.4.1 To infinity ‚ôæ‚¨Öüöóüí®
**... And beyond!**

Although we cannot [divide](https://www.youtube.com/watch?v=NKmGVE85GUU) by [zero](https://www.youtube.com/watch?v=BRRolKTlF6Q), because it's kind of nonsensical (see <a href="#apdx1">**appendix 1**</a>),
we can see that as $x \to 0$ from 1, $\frac{1}{x}$ gets bigger and <font size=4>bigger</font> <font size=3> and </font><font size=6>bigger.</font>

<br>

Since there are `infinitely` many numbers between `1` and `0` (and between any different numbers), $\frac{1}{x}$ increases infinitely as $x \to 0$.


But just as $x$ can never get to zero (otherwise the universe breaks down), $\frac{1}{x}$ can never get to a number, "infinity." `Infinity` is not a quantity, but the idea of a never-ending process.


<a name="ch2.4.2"></a>
## 2.4.2 Infinitesimals üî¨ü§è

We saw that, as $\Delta{x} \to 0$, $\frac{1}{\Delta{x}} \to \infty$

But what about, as $\Delta{x} \to \infty$, $\frac{1}{\Delta{x}} \to ?$

I don't know ü§∑‚Äç‚ôÇÔ∏è. But one possible answer is:

$$
\Delta{x} \to \infty, \,\, \frac{1}{\Delta{x}} \to \epsilon
$$

We call $\epsilon$ an [infinitesimal](https://www.youtube.com/watch?v=WYijIV5JrKg). The `infinitesimal` is also a never-ending process. It's a process where a quantity between `1` and `0` becomes **smaller and <font size=1>smaller and</font> <font size=0>...</font>**. Whereas `infinity` is a process where a quantity becomes larger and larger.

See **<a href="#adpx2">appendix 2</a>** for some fun facts about infinitesimals.

```python
# Ch 2.4 Question 2.

def zeno_dichotomy(n):
    
    i = 0
    
    try:
        while n!=0:
            n=n/2
            i+=1

    except Exception as e:
        print(e)
    
    return i, n
```

# Ch 2.5: Limits üõë‚¨Öüö∂‚Äç‚ôÇÔ∏èüö∂‚Äç‚ôÄÔ∏è

So here's an interesting question: as we saw in our challenges, why doesn't the function:

$$\frac{f(x+\Delta{x}) - f(\Delta{x})}{\Delta{x}}$$

blow up as we made $\Delta{x} \to 0$? As the `denominator` becomes infinitely small, should the value of the fraction become infintely larger?

Not only does the function not go to infinity, it seems to approach a *`single, finite number`*. ü§î

One might even say that as the input progresses `infinitely`, the value of the function is **`limited`üòâ** by this `finite` number. 

To explain this, we should examine the concept of ***`limits`***:

<a name="ch2.5.1"></a>
## 2.5.1: A definition of limits

From Wikipedia (a little lazy, I know üòú):

$\qquad$"*A limit is the value that a function (or sequence) "approaches" as the input "approaches" some value.*"


### Limit of a variable

Say we have a `variable` quantity $h$:

$$
\begin{align}
h, \text{ our variable}\\
\end{align}
$$

And a `constant` quantity $c$:

$$
\begin{align}
c &= 2, \text{ for example}\\
\end{align}
$$

We say that as $h \to c$, the **`limit`** of $h$ ***`equals`*** $c$:
$$
\begin{align}
\lim_{h \to c}\,\, h &= c \\
\lim_{h \to 2}\,\, h &= 2 \\
\lim_{h \to 0}\,\, h &= 0 \\
\lim_{h \to \infty}\,\, h &= \infty \\
\end{align}
$$

### Limit of a function

Say we have a quantity that is a `function` of $h$, for example:

$$
\begin{align}
f(h) = \frac{1}{h}
\end{align}
$$

We can take the `limit` of this function as $h \to c$:

$$
\begin{align}
\lim_{h \to c} \frac{1}{h} &= \frac{1}{c} \\
\lim_{h \to 2} \frac{1}{h} &= \frac{1}{2} \\
\lim_{h \to \infty} \frac{1}{h} &= 0 \\
\lim_{h \to 0} \frac{1}{h} &= \pm \infty \\
\end{align}
$$

### Approaching Zero Positively or Negatively

For that last example, we say that as $h \to 0$, the value of the function approaches `positive` ***or*** `negative` infinity:

$$
\lim_{h \to 0} \frac{1}{h} = \pm \infty
$$

This is because whether the function is positive or negative depends on the ***`direction`*** from which $h$ approaches zero.


Say we pick a random value for $h$, and then have $h \to 0$. If $h$ is negative:

$$
\begin{align}
h < 0 \\
h = -1: \frac{1}{h} &= -1 \\
h = -0.1: \frac{1}{h} &= -10 \\
h = -0.00001: \frac{1}{h} &= -100,000 \\
& \quad \vdots \\
h \to 0: \frac{1}{h} &= -\infty \\
\end{align}
$$

Here, we say that $h$ approaches zero **`negatively`**, and write:

$$
\lim_{h \to 0^{-}} \frac{1}{h} = -\infty
$$


Similarly, if we can begin with choosing $h$ as a positive number. As $h \to 0$, we say that $h$ approaches zero **`positively`**, and write:
$$
\lim_{h \to 0^{+}} \frac{1}{h} = -+\infty
$$

<a name="ch2.5.2"></a>
## 2.5.2 Examples of limits

It's interesting to think about what value a function will become as we take the limit for one of its variable. However, this value is not always obvious, especially if the function is more complicated or nuanced.

For now, we'll look at some more simple examples to get the hang of limits.

There is a great [Wikipedia](https://en.wikipedia.org/wiki/List_of_limits) article with a list limits of common functions.

Here is a brief list of the limits of `functions` with different arithmetic operations. In these examples, $a, b$ are independent `variables` of $h$, and $c$ is a `constant`. You can quickly glance at these and move on to the `challenges`, where you'll solve the limits of some more interesting functions:

$$
\begin{align}
\\
\text{Addition: } \qquad&
\\
\lim_{h \to c}\, a + h &= a + c \\
\lim_{h \to 0}\, a + h &= a \\
\lim_{h \to \pm \infty}\, a + h &= \pm \infty \\
\\
\text{Multiplication: } \qquad&
\\
\lim_{h \to c}\, \frac{ah}{b} &= \frac{ac}{b} \\
\lim_{h \to 0}\, \frac{ah}{b} &= 0 \\
\lim_{h \to \pm \infty}\, \frac{ah}{b} &= \pm \infty \\
\\
\text{Inverse: } \qquad&
\\
\lim_{h \to \pm c}\, \frac{a}{bh} &= \frac{a}{bc} \\
\lim_{h \to 0^{\pm}}\, \frac{a}{bh} &= \pm \infty \\
\lim_{h \to \pm \infty}\, \frac{a}{bh} &= 0 \\
\\
\text{Exponents: } \qquad&
\\

\text{if } a > b: \quad&\\
\lim_{h \to 0}\, \frac{h^a}{h^b} &= 0 \\
\lim_{h \to \pm \infty}\, \frac{h^a}{h^b} &= \pm \infty \\

\lim_{h \to 0^{\pm}}\, \frac{h^a}{h^b} &= \pm \infty \\
\lim_{h \to \pm \infty}\, \frac{h^a}{h^b} &= 0 \\

# Ch 2.6: Differentials and Derivatives


<a name="ch2.6.1"></a>
## 2.6.1: Differentials
We call the limit of **`the change`** in a function $f(x)$:
1. as a result of a change in $x$, labeled the quantity $\Delta{x}$,
2. as $\Delta{x} \to \infty$

the **`differential`** of $f(x)$:

$$
\begin{align}
y &= f(x) \\
\\
y_1 &= f(x + \Delta{x})\\
\\
y_1 - y  &= f(x + \Delta{x}) - f(x) \\
\\
\Delta{y}  &= f(x + \Delta{x}) - f(x) \\
\\
\text{Taking the limit, we get:}\\
\\
\lim_{\Delta{x} \to 0}\Delta{y}  &= \lim_{\Delta{x} \to 0}f(x + \Delta{x}) - f(x)
\\
\\
\text{We shorten this expression by writing:}\\
=> dy &= df \\
\\
\end{align}
$$

<a name="ch2.6.2"></a>
## 2.6.2 Derivatives
We call the limit of **`the slope`** of a function $f(x)$ at a point:
1. as a result of a change in $x$, labeled the quantity $\Delta{x}$,
2. as $\Delta{x} \to \infty$

the **`derivative`** of $f(x)$:

$$
\begin{align}
\lim_{\Delta{x} \to 0}\frac{\Delta{y}}{\Delta{x}}
&=\lim_{\Delta{x} \to 0}\frac{f(x+\Delta{x})-f(x)}{\Delta{x}} \\
\\
\text{We shorten this expression by:}&\\
\\
\text{1. } & \text{removing the notation for the limits, and;} \\
\text{2. } & \text{changing the notation for the "$\Delta$" variables that the limit applies to} \\
\\
\text{And we write:}&\\
\lim_{\Delta{x} \to 0}\frac{\Delta{y}}{\Delta{x}}
&= \frac{dy}{dx} =\frac{f(x+dx)-f(x)}{dx} \\
\end{align}
$$

### Note on Notation

There are some choices of notation for expressing the derivative:

$$
\begin{align}
\frac{dy}{dx} &= \frac{df(x)}{dx} = \frac{df}{dx} \quad\text{Leibnez's Notation} \\
\\
y' &= f'(x) = f' \quad\text{Lagrange's Notation}\\
\end{align}
$$

All the notation `Leibnez` and `Lagrange notations` above describe the same thing. There are more details on derivative notations in **<a href="#apdx3">appendix 3</a>**.

## 2.7.1: Overview of formulation

Let's quickly reformulate the derivative, applying the slope of a line and limits:

$$
\require{cancel}
\begin{align}
\\
\text{We have some function: }\\
y &= f(x) \\
\\
\\
\text{Take two points } x_1, x_2 &\text{ and get the difference after plugging into} f(x):\\
\\
y_1 - y_0 &= f(x_1)- f(x_0) \\
\\
\\
\text{Get the slope of change:}\\
\\
\frac{y_1 - y_0}{x_1-x_0} &= \frac{f(x_1)- f(x_0)}{x_1-x_0} \\
\\
\frac{\Delta{y}}{x_0+\Delta{x}-x_0} &= \frac{f(x_0+\Delta{x})- f(x_0)}{x_0+\Delta{x}-x_0} \\
\\
\frac{\Delta{y}}{\Delta{x}} &= \frac{f(x_0+\Delta{x})- f(x_0)}{\Delta{x}} \\
\\
\\
\text{Take the limit as} \Delta{x} \to 0\\
\\
\lim_{\Delta{x \to 0}}\frac{\Delta{y}}{\Delta{x}} &= \lim_{\Delta{x \to 0}}\frac{f(x_0+\Delta{x})- f(x_0)}{\Delta{x}} \\
\\
\\
\text{Since } x_{0} \text{ is unspecified, we can} & \text{ apply the formulation to any } x:\\
\\
=>\lim_{\Delta{x \to 0}}\frac{\Delta{y}}{\Delta{x}} &= \lim_{\Delta{x \to 0}}\frac{f(x+\Delta{x})- f(x)}{\Delta{x}} \\
\\
\\
\text{For brevity, we convert the Leibniz notation} & \text{ to remove the limits:} \\
\\
=>\frac{d{y}}{d{x}}  &= \frac{f(x+dx)- f(x)}{dx}\\
\\
\end{align}
$$

<a name="ch2.7.2"></a>
## 2.7.2: The function of a derivative üïµÔ∏è‚Äç‚ôÄÔ∏è

Not only can we use the formulation to solve for `specific values` of the derivative, we can also analytically solve for a **`function`** that, given a **`domain`**, describes the **`range`** of the derivative.

Let's plug in our `example function` from *`Ch 2.1.`*, $f(x) =  \frac{x^3}{3!}$ and see what happens:

$$
\begin{align}
\\
\frac{d{y}}{d{x}}  &= \frac{f(x+dx)- f(x)}{dx}, \, f(x) =  \frac{x^3}{3!}\\
\\
\frac{d{y}}{d{x}} &= \frac{(x+dx)^3- x^3}{3!dx} \\
\\
&= \frac{(x+dx)(x+dx)(x+dx)-x^3}{3!dx} \\
\\
&= \frac{(x^3+3x^2dx+3xdx^2+dx^3)-x^3}{3!dx} \\
\\
&= \frac{x^3}{3!dx}
+\frac{3x^2dx}{dx}
+\frac{3xdx^2}{3!dx}
+\frac{dx^3}{3!dx}
-\frac{x^3}{3!dx}\\
\\
\text{Cancel some terms and factors:} & \\\\
&= \cancel{\frac{x^3}{3!dx}}
+\frac{\cancel{3}x^2dx}{\cancel{3!}2!dx}
+\frac{3xdx^2}{3!dx}
+\frac{dx^3}{3!dx}
-\cancel{\frac{x^3}{3!dx}}\\
\\
\text{Since $dx \to 0$, the terms $\to 0$ if } & \text{it has a higher order of $dx$ in the numerator than denominator:} \\\\
&=\frac{x^2dx}{2!dx}
+\cancel{\frac{3xdx^2}{3!dx}}
+\cancel{\frac{dx^3}{3!dx}} \\
\\
&= \frac{x^2dx}{2!dx}
\\\\
&= \frac{x^2}{2!}\\
\end{align}
$$


We now have a **`function`** for the **`derivative`** of `f(x)`:

$$
\begin{align}
\text{Original Function:}\quad y &= \frac{x^3}{3!}\\
\text{Derivative function:}\quad \frac{d{y}}{d{x}} &= \frac{x^2}{2!} \\
\end{align}
$$

### <a name="apdx1">Appendix I:</a> Dividing by Zero

Suppose we define `division` as separating a number into equal additive groups:

i.e.:

* Divide `12 into 5` groups = `2.4 + 2.4 + 2.4 + 2.4 + 2.4`, therefore `12 / 5 = 2.4`


* Divide `x into 100` groups = `y + y + y...`, with `100 y's`, therefore `x / 100 = y`

Following this logic, consider what it means to divide a number by zero.


* Divide `1 into 0` groups = `...`, with `0.. wha..? üòê`


The phrase, separate *`something`* into *`no groups`*, doesn't really make sense. Just like you can't make something out of nothing, you can't create "nothingness" out of "something" either. That "something" has to exist somewhere.

You can split an atom into smaller and smaller parts, you cannot split an atom into nothing. The same is true, *I think*, of numbers.

### <a name="apdx2">Appendix II:</a> Fun facts about Infinitesimals

1. Back in the days of Newton and Leibnez, calculus was referred to as the **`calculus of infinitesimals.`** Interesting...? ü§∑‚Äç‚ôÄÔ∏è



2. Formal definition of `infinitesimal`:

    ***An infinitesimal is a number whose magnitude exceeds zero but somehow fails to exceed any finite, positive number.***

[*Joel A. Tropp - Infinitesimals: History & Application*](http://users.cms.caltech.edu/~jtropp/papers/Tro99-Infinitesimals.pdf)


3. Courtesy of [the same guy](http://users.cms.caltech.edu/~jtropp/papers/Tro99-Infinitesimals.pdf) who provided the definition above, `infinitesimals` have the such operations with the *`hyperreal number`* according to another guy named [Robert Goldblatt](https://homepages.ecs.vuw.ac.nz/~rob/):

<br>
$$
\begin{align}
\text{Let: }
\epsilon, \delta \text{ be infinitesimal, } &
b, c \text{ be appreciable,} 
\text{ and } N, M \text{ unlimited:} \\
\\
\text{Sums:} \\
&\epsilon + \delta \text{ is infinitesimal;} \\
&b + \epsilon \text{ is appreciable;} \\
&b + c \text{ is limited (possibly infinitesimal);} \\
&N + \epsilon \text{ and } N + b \text{ are unlimited.} \\
\\
\text{Products:}&  \\
&\epsilon ¬∑ \delta \text{ and } \epsilon ¬∑ b \text{ are infinitesimal;} \\
&b ¬∑ c \text{ is appreciable;} \\
&b ¬∑ N \text{ and } N ¬∑ M \text{ are unlimited.} \\
\\
\text{ Reciprocals:}& \\
&\frac{1}{\epsilon} \text{ is unlimited if } \epsilon \ne 0; \\
&\frac{1}{b} \text{ is appreciable;} \\
&\frac{1}{N} \text{ is infinitesimal.} \\
\\
\text{Roots:}& \\
\text{For } & n ‚àà N, \\
&\text{if } \epsilon > 0, {\epsilon}^{1/n}  \text{ is infinitesimal;} \\
&\text{if } b > 0, b^{1/3} \text{ is appreciable;} \\
&\text{if } N > 0, N^{1/3} \text{ is unlimited.} \\
\\
\text{Indeterminate Forms:}& \\
&\frac{\epsilon}{\delta}, \frac{N}{M}, \epsilon ¬∑ N, N + M. \\
\end{align}
$$

Euler's notation is also a little uncommon.

#### Taking our previous example of $x(t)$:

<br><br>
$$
\begin{align}
\text{Leibniz's Notation:} \qquad\qquad\qquad& \\
\\
1^{st} \text{derivative:}\qquad & \frac{d x(t)}{dt} = v(t) \\ 
\\
2^{nd} \text{derivative:} \qquad & \frac{d^{2} x(t)}{dt^{2}} = \frac{d v(t)}{dt} = a(t)\\ 
\\
3^{rd} \text{derivative:} \qquad & \frac{d^{3} x(t)}{dt^{3}} = \frac{d^{2} v(t)}{dt^{2}} = \frac{d a(t)}{dt} = ...\\ 
\vdots \qquad \qquad & \qquad \vdots \\
\\\\
\text{Lagrange's Notation:} \qquad\qquad\qquad& \\
\\
1^{st} \text{derivative:} \qquad & x'(t) = v(t) \\ 
\\
2^{nd} \text{derivative:} \qquad & x''(t) = x^{(2)}(t) = v'(t) = a(t)\\ 
\\
3^{rd} \text{derivative:} \qquad & x'''(t) = x^{(3)}(t) = v''(t) = v^{(2)}(t) = a'(t) = ...\\ 
\vdots \qquad \qquad & \qquad \vdots \\
\\\\
\text{Newton's Notation:} \qquad\qquad\qquad& \\
\\
1^{st} \text{derivative:} \qquad & \dot{x}(t) = v(t) \\ 
\\
2^{nd} \text{derivative:} \qquad & \ddot{x}(t) =  \dot{v}(t) = a(t)\\ 
\\
3^{rd} \text{derivative:} \qquad & \dddot{x}(t) = \overset{.}{\ddot{x}} = \ddot{v}(t) = \dot{a}(t) = ...\\
\\
4^{th} \text{derivative:} \qquad & \ddddot{x}(t) = \overset{4}{\dot{x}} = \dddot{v}(t) = \ddot{a}(t) = ...\\
\vdots \qquad \qquad & \qquad \vdots \\
\\\\
\text{Euler's Notation:} \qquad\qquad\qquad& \\
\\
1^{st} \text{derivative:} \qquad & (Dx)(t) = v(t) \\ 
\\
2^{nd} \text{derivative:} \qquad & D^{2}x = (Dv)(t) = a(t)\\ 
\\
3^{rd} \text{derivative:} \qquad & D^{3}x = D^{2}v = (Da)(t)\\ 
\vdots \qquad \qquad & \qquad \vdots \\
\end{align}
$$
