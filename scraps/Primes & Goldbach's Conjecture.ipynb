### Primes

```python
_primes = [2]

def index_prime(n):
    '''
    Get n-th prime (starting n=0).
    '''
    i = _primes[-1] # current largest prime
    while n >= len(_primes):
        i += 1
        if all(i % p != 0 for p in _primes):
            _primes.append(i)
        
    return _primes[n]


def largest_prime(n, index=False):
    '''
    Get the largest prime <= n.
    '''
    
    # start at largest prime in list
    i = _primes[-1]
    
    # keep finding _primes until looked through to n
    while n >= i:
        if all(i % p != 0 for p in _primes):
            _primes.append(i)
        i+=1
    
    # if n closer to the top of list of _primes
    if (_primes[-1] - n) < (n - _primes[0]):
        for p in reversed(_primes):
            if p <= n:
                return p
            
    else: # if closer to the bottom
        for i, p in enumerate(_primes):    
            if p > n:
                return _primes[i-1]


def list_prime(n):
    '''
    List all primes <= n
    '''
    p = largest_prime(n)
    i = _primes.index(p)
    
    return _primes[:i]


def is_prime(n):
    '''
    Check if n is prime
    '''
    return largest_prime(n) == n
```

### Factoring

```python
def factor(number):

    factors = []

    i = 0
    # while factors are remaining
    while number > 1:

        p = index_prime(i)

        # while divisible
        while number % p==0:
            factors.append(p) # add factor
            number /= p # divide number by factor

        i += 1

    return factors
```

```python
def product(*args):
    
    res = 1
    if len(args)==1:
        args = args[0]
    
    for arg in args:
        res *= arg
    return res
```

```python
n = 4129
factors = factor(n)
print(product(factors) == n)
print(n, factors)
```

### Goldbach's Conjecture:

All even integers other than 2 are the sum of 2 primes.

```python
%%time

# for all primes under limit
limit = 100_000
primes = list_prime(limit)

# the sum of the primes
prime_sums = dict()
for i, x in enumerate(primes):
    for j, y in enumerate(primes[i:]):
        prime_sums[x+y] = (x, y)
        
        
# creates all even numbers
result = {i: (i in prime_sums) for i in prime_sums if i%2==0}

conjecture_holds = all((result.values()))

print(f'Conjecture holds: {conjecture_holds} | Numbers: {max(prime_sums)}\n')
```

```python
%%time
for i in result:
    print(i, prime_sums[i])
```
