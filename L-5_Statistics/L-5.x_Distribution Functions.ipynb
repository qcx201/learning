```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
def plot_pdf(f,
             x_range=None,
             results=False,
             threshold=0.0001,
             *args, **kwargs):
    
    if x_range:
        start, stop = x_range
    else:
        start = -100
        while True:
            if f(start) > threshold:
                start = 1.25 * (start - 0.5)
                break
            else:
                start += 0.1
                
        stop = 100
        while True:
            if f(stop) > threshold:
                stop = 1.25 * (stop + 0.5)
                break
            else:
                stop -= 0.1
        
    
    X = np.linspace(start, stop, 1000)
    P = np.array([f(x, *args, **kwargs) for x in X])
    
    plt.plot(X, P, lw=5, alpha=0.6)

    plt.title(f'"{f.__name__.title()}" PMF',
              fontsize=14, fontweight='bold')
    
    plt.xlabel('Random variable $x$')
    plt.ylabel('Probability $p(x)$')
    
    plt.show()
    
    if results:
        return X, P
```

```python
def uniform(x):
    
    if 0 < x <= 1:
        return 1
    else:
        return 0

def exp(x, λ=0.5):
    if x <= 0 :
        return 0
    else:
        return 1 - np.exp(-λ*x)
    
def normal(x, μ=0, σ=1):
    return (
        np.exp(-( (x - μ)**2 )/ ( 2 * σ**2 ) ) /
        (σ * np.sqrt(2*np.pi))
    )
```

```python
res = plot_pdf(uniform)
```

```python
res = plot_pdf(exp)
```

```python
res = plot_pdf(normal)
```
